```python
import numpy as np
import pandas as pd
```

```python
# importing sample of global supertore dataset
df = pd.read_csv('https://raw.githubusercontent.com/yannie28/Global-Superstore/master/Global_Superstore_100.csv')
df.head(3)
```

## Assign method to create new columns to the dataframe

Note: assign creates a copy of the dataset, and changes on newdf doesn't effect df ie .copy() applied by default.  
Can create multiple columns at once  
Useful for method chaining

```python
new_df = df.assign(month_order_date = lambda x:pd.to_datetime(x['Order Date']).dt.month,
                   year_order_date = lambda x:pd.to_datetime(x['Order Date']).dt.year,
                   msp = df['Sales']/df['Discount'],
                   msp_final = lambda x: np.where(x['msp']==np.inf,x['Sales'],x['msp']),
                   unit_price = lambda x:np.round(x['msp_final']/x['Quantity'],2),
                   ).drop(columns=['msp']).rename(columns={'msp_final':'msp'})
```
